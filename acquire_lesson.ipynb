## NY Times Website

```python
response = requests.get('https://www.nytimes.com/')
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
soup.find_all('h3')
```

```python
soup.find_all('h3', class_='indicate-hover')
```

```python
[el.text for el in soup.find_all('h3', class_='indicate-hover')]
```

```python
soup.find_all('h2')
```

```python
soup.find_all('h3')
```

```python
soup.find_all('a')
```

```python
article_links = soup.find_all('a', class_='css-9mylee')
article_links
```

```python
article_links[0]['href']
```

```python
for element in article_links:
    print(element['href'])
```

```python
links_list = []

for element in article_links:
    links_list.append(element['href'])
```

```python
links_list
```

```python
url = links_list[0]

response = requests.get(url)
response
```

## How to use headers

```python
url = 'https://codeup.com/blog/'
response = requests.get(url)
```

```python
response
```

```python
url = 'https://codeup.com/blog/'
headers = {'User-Agent': 'Codeup Data Science'}
response = requests.get(url, headers=headers)
```

```python
response
```
